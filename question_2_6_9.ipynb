```python
import matplotlib.pyplot as plt

# Data defined directly in the script
counties = [
    'Chambers County', 'Walker County', 'Lowndes County', 'Dallas County',
    'Autauga County', 'Wilcox County', 'Lamar County', 'Covington County',
    'Elmore County', 'Bullock County'
]
avg_annual_count = [240, 447, 71, 258, 304, 68, 100, 251, 436, 60]  # Population sizes
incidence_rate_per_100k = [522.8, 515.8, 513.3, 510, 495.6, 488, 487.1, 486, 485.8, 481.3]  # Incidence rates

# --- Scatter Plot ---
plt.figure(figsize=(10, 6))

# Create scatter plot
plt.scatter(avg_annual_count, incidence_rate_per_100k, color='purple', s=100, alpha=0.7)

# Add labels and title
plt.title('Scatter Plot: Cancer Incidence Rates vs Population Size')
plt.xlabel('Average Annual Count (Population Size)')
plt.ylabel('Incidence Rate per 100k')

# Annotate points with county names
for i, county in enumerate(counties):
    plt.text(avg_annual_count[i], incidence_rate_per_100k[i], county, fontsize=8, ha='right', alpha=0.7)

# Grid for better readability
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```
