```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for the first 25 counties
data = {
    'county': [
        'Chambers County', 'Walker County', 'Lowndes County', 'Dallas County', 'Wilcox County', 'Lamar County',
        'Covington County', 'Elmore County', 'Russell County', 'Etowah County', 'Bibb County', 'Tallapoosa County',
        'Clarke County', 'Limestone County', 'Macon County', 'Marion County', 'Crenshaw County', 'Greene County',
        'Morgan County', 'Clay County', 'Jackson County', 'Sumter County', 'Jefferson County', 'Marengo County',
        'Pike County', 'Conecuh County', 'Henry County', 'Calhoun County'
    ],
    'incidence_rate_per_100k_low_95': [
        492.7, 494.0, 459.8, 481.6, 470.6, 435.7, 443.9, 458.4, 465.1, 427.3,
        455.7, 461.9, 441.2, 449.5, 454.9, 438.5, 453.1, 432.7, 441.5, 441.1,
        424.6, 411.5, 450.9, 421.0, 417.7, 434.6, 407.0, 448.6
    ],
    'incidence_rate_per_100k_high_95': [
        554.5, 538.4, 571.8, 539.7, 521.5, 545.2, 533.8, 515.0, 507.2, 540.6,
        505.7, 496.1, 517.8, 503.0, 493.8, 507.7, 491.7, 512.7, 502.7, 501.4,
        517.5, 531.2, 482.9, 513.6, 506.0, 481.1, 510.2, 462.5
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot dot plot
fig, ax = plt.subplots(figsize=(16, 8))

# Plot lower bounds (95% confidence intervals lower)
ax.scatter(df['incidence_rate_per_100k_low_95'], df['county'], color='skyblue', label='95% CI Lower Bound')

# Plot upper bounds (95% confidence intervals upper)
ax.scatter(df['incidence_rate_per_100k_high_95'], df['county'], color='salmon', label='95% CI Upper Bound')

# Add labels and title
ax.set_xlabel('Incidence Rate per 100,000')
ax.set_ylabel('County')
ax.set_title('Cancer Incidence Rates (95% Confidence Intervals) by County')

# Add legend
ax.legend()

# Show grid for better visualization
ax.grid(True, linestyle='--', alpha=0.7)

# Adjust layout
fig.tight_layout()

# Show plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data defined directly in the script (First 25 counties)
counties = [
    'Chambers County', 'Walker County', 'Lowndes County', 'Dallas County', 'Autauga County',
    'Wilcox County', 'Lamar County', 'Covington County', 'Elmore County', 'Bullock County',
    'Russell County', 'Etowah County', 'Bibb County', 'Tallapoosa County', 'St. Clair County',
    'Clarke County', 'Limestone County', 'Macon County', 'Marion County', 'Lawrence County',
    'Crenshaw County', 'Greene County', 'Morgan County', 'Clay County', 'Cleburne County'
]

avg_annual_count = [
    240, 447, 71, 258, 304, 68, 100, 251, 436, 60,
    255, 345, 456, 389, 398, 224, 311, 429, 360, 478,
    196, 344, 259, 373, 467, 414, 315
]

incidence_rate_per_100k = [
    522.8, 515.8, 513.3, 510, 495.6, 488, 487.1, 486, 485.8, 481.3,
    480.2, 478.7, 478.3, 475.6, 474, 472.1, 472.1, 471.3, 471.3, 470.5,
    469.2, 468.2, 466.7, 465.4, 460.1, 457.4, 456.7
]

# --- Scatter Plot ---
plt.figure(figsize=(12, 8))

# Create scatter plot
plt.scatter(avg_annual_count, incidence_rate_per_100k, color='purple', s=100)

# Add labels and title
plt.title('Scatter Plot: Cancer Incidence Rates vs Population Size (First 25 Counties)')
plt.xlabel('Average Annual Count (Population Size)')
plt.ylabel('Incidence Rate per 100k')

# Annotate points with county names
for i, county in enumerate(counties):
    plt.text(avg_annual_count[i], incidence_rate_per_100k[i], county, fontsize=9, ha='right')

# Grid for better readability
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```
